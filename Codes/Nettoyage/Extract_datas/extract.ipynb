```python
import pandas as pd
from reportlab.lib.pagesizes import A4
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from xml.sax.saxutils import escape  # Import correct de la fonction d'échappement
import os

def generate_text_pdf(df, output_path='dataframe_analysis.pdf'):
    """
    Génère un PDF unique avec toutes les lignes du DataFrame présentées sous forme de texte structuré.
    
    :param df: pandas DataFrame préparé
    :param output_path: Chemin de sortie pour le PDF
    """
    # Créer le document PDF
    doc = SimpleDocTemplate(
        output_path,
        pagesize=A4,
        rightMargin=30, leftMargin=30,
        topMargin=30, bottomMargin=18
    )
    
    elements = []
    styles = getSampleStyleSheet()
    heading_style = styles['Heading1']
    subheading_style = styles['Heading2']
    field_style = ParagraphStyle(
        'FieldStyle',
        parent=styles['Normal'],
        fontName='Helvetica',
        fontSize=10,
        leading=12,
        spaceAfter=4
    )
    
    # Titre du document
    title = Paragraph("Analyse des Données - DataFrame", heading_style)
    elements.append(title)
    elements.append(Spacer(1, 12))
    
    # Définir les champs à inclure
    fields = ['id', 'nom', 'date_ouverture', 'date_fermeture', 'website', 'email',
              'adresse', 'horaires', 'presentation', 'responsables', 'historique',
              'activites', 'expositions', 'publications', 'events', 'residences',
              'workshop', 'archive', 'librairie', 'coworking', 'education',
              'nb_expos_an', 'nb_artistes_an', 'partenaires', 'created_at',
              'updated_at', 'slug', 'pays', 'ville', 'latitude', 'longitude',
              'adresse_complete', 'codepostal', 'telephone', 'visite', 'user_id',
              'réponse1', 'réponse2']
    
    # Parcourir chaque ligne du DataFrame
    for index, row in df.iterrows():
        # Ajouter un séparateur pour chaque enregistrement
        if index != 0:
            elements.append(Spacer(1, 12))
            elements.append(PageBreak())
            elements.append(Spacer(1, 12))
        
        # Titre de l'enregistrement
        record_title = Paragraph(f"Enregistrement {index + 1}: {escape(row['nom'])}", subheading_style)
        elements.append(record_title)
        elements.append(Spacer(1, 6))
        
        # Ajouter chaque champ et sa valeur
        for field in fields:
            field_name = field.capitalize().replace('_', ' ')
            field_value = escape(str(row[field]))  # Échappement des caractères spéciaux
            # Format: Champ: Valeur
            text = f"<b>{field_name}:</b> {field_value}"
            paragraph = Paragraph(text, field_style)
            elements.append(paragraph)
    
    # Construire le PDF
    doc.build(elements)

if __name__ == '__main__':
    # Remplacez cette partie par votre méthode de chargement réelle du DataFrame
    # Par exemple, si votre DataFrame est déjà chargé dans une variable `df`, utilisez-le directement
    
    # Exemple d'importation d'un fichier CSV (décommentez et modifiez le chemin si nécessaire)
    # df = pd.read_csv('votre_fichier.csv')
    
    # Assurez-vous que votre DataFrame `df` est déjà chargé ici
    # Par exemple :
    # df = pd.read_excel('votre_fichier.xlsx') ou toute autre méthode de chargement
    
    # Vérifiez que `df` est défini avant de continuer
    try:
        df
    except NameError:
        print("Veuillez charger votre DataFrame avant d'exécuter ce script.")
        exit(1)
    
    # Remplacer les NaN : numériques par 0, non numériques par "Information Manquante"
    numeric_cols = df.select_dtypes(include=['number']).columns
    non_numeric_cols = df.select_dtypes(exclude=['number']).columns
    df[numeric_cols] = df[numeric_cols].fillna(0)
    df[non_numeric_cols] = df[non_numeric_cols].fillna("Information Manquante")
    
    # Générer le PDF
    generate_text_pdf(df, output_path='dataframe_analysis.pdf')
    
    print("PDF généré avec succès : dataframe_analysis.pdf")
```
